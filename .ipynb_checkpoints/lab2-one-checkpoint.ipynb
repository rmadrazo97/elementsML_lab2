### Linear Regression Implementation

### Simple Linear Regression

```python
data.shape
```

```python
y = data["chance_admit"].to_numpy() # get y data
y = y.reshape(-1,1) # convert y to matrix
x = data["gre_score"].to_numpy() # get x data
x = x.reshape(-1,1) # convert x to matrix
unos = np.ones(x.shape) # array of ones "unos"
x = np.concatenate([unos, x],1) # concatenate x and unos to fill as [1,x]
```

```python
plt.scatter(data.gre_score, data.chance_admit) # this is what we are trying to evaluate
```

```python
def hola():
    return len(y), y.shape
hola()
```

```python
def hola2():
    return len(x), x.shape
hola2()
```

```python
# setting hyperparams (learning rate & number of iterations)
a = 0.000000001
it = 1000000000
# thetas: pendiente (m) e intercepto (b) con eje y -> y = mx+b
ths = np.array([[1.0,1.0]])

def cost(x,y,thetas):
    two_m = 2*len(x) # 1/2m
    hip = np.matmul(x, thetas[0]) # hypothesis
    to_sum = np.power(hip - y, 2) # inner section to sum
    return np.sum(to_sum)/two_m # final result returns the difference between our prediction and real values
```

```python
c = cost(x,y,ths)
```

```python
def grd_dsc(x,y,theta,a,it):
    for i in range(it):
        theta = theta - (a/len(x)) * np.sum((x @ theta.T - y) * x, axis=0)
        cost1 = cost(x, y, theta)
        # if i % 10 == 0: # just look at cost every ten loops for debugging
        #     print(cost)
        return (theta, cost)
```

```python
g = grd_dsc(x,y,ths,a,it)
```

```python
print("Cost: ",c)
```

```python
print("Gradient: ",g)
```

```python
#plot resulting linear regression

plt.scatter(data.gre_score, data.chance_admit)
axes = plt.gca() # reference to axes
x_values = np.array(axes.get_xlim())  
y_values = g[0][0][0] + g[0][0][1]* x_values # y = b+mx (line eq)
plt.plot(x_values, y_values, '--')
```
