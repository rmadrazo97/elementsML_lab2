## Default Hyperparams & Thetas

```python
# configurating hyperparams and thetas
thetas = np.zeros([1,7])
a = 0.00001
it = 1000
```

```python
def cost1(x,y,thetas):
    m = 2 * len(x)
    hip = x @ thetas.T # using @ as matmul
    to_sum = np.power(hip - y,2)
    return np.sum(to_sum)/m
```

```python
cost1(x,y,thetas)
```

```python
# gradient descent function
def grdDsc(x,y,thetas,a,ite):
    cost = np.zeros(ite)
    for i in range(ite):
        hip = x @ thetas.T
        to_sum = x * (hip - y)
        thetas = thetas - (a/len(x)) * np.sum(to_sum, axis=0)
        # calling cost function
        cost[i] = cost1(x, y, thetas)
        
    # return cost and thethas together
    return thetas,cost
```

## calling function grdDsc Gradient Descent with different methods of normalization

```python
def call_grdDsc(x,y,thethas,a,it):
    #running the gd and cost function
    g,cost = grdDsc(x,y,thetas,a,it)
    #print("Gradient Result : ", g)
    #print("Costs result: ", cost)
    initialCost = cost1(x,y,thetas)
    #print("Initial Cost:",initialCost)
    change = (abs(cost[-1]-initialCost)/initialCost) * 100
    
    # plotting cost
    fig, ax = plt.subplots()  
    ax.plot(np.arange(it), cost, 'r')  
    ax.set_xlabel('Iterations')  
    ax.set_ylabel('Cost')  
    ax.set_title('Error vs. Training iterations') 
    print("Initial Cost:",initialCost,"- Final Cost:",cost[-1], " - Improvement:",change,"%")
    return g
```

### Multiple Linear Regression - Regularized Version

```python
def cost2(x,y,thetas,lamb):
    m = 2 * len(x)
    hip = x @ thetas.T # using @ as matmul
    to_sum = np.power(hip - y,2)
    return ( np.sum(to_sum)/m ) + (lamb * np.sum(thetas)) / m
```

```python
cost2(x,y,thetas,0.5)
```

```python
# gradient descent function
def grdDsc2(x,y,thetas,a,ite, lamb):
    costs = np.zeros(ite)
    for i in range(ite):
        m = len(x)
        hip = x @ thetas.T
        to_sum = x * (hip - y)
        thetas = thetas - (a/m) * np.sum(to_sum, axis=0) + (lamb * thetas)/m
        # calling cost function
        costs[i] = cost2(x, y, thetas, lamb)
        
    # return cost and thethas together
    return thetas,costs

#running the grdDsc and cost function from "iterable model"
g2,costs2 = grdDsc2(x,y,thetas,a,it,0.5)
#print("Gradient Result : ", g)
#print("Costs result: ", cost)

#initial cost
initialCost2 = cost2(x,y,thetas,0.5)
#print("Initial Cost:",initialCost)

change2 = (abs(costs2[-1]-initialCost2)/initialCost2)*100
```

```python
print("Initial Cost:",initialCost2,"- Final Cost:",costs2[-1], " - Change:",change2,"%")
```

```python
# plotting cost
fig, ax = plt.subplots()  
ax.plot(np.arange(it), costs2, 'r')  
ax.set_xlabel('Iterations')  
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Training iterations') 
```
